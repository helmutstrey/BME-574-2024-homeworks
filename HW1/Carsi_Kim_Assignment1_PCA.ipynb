```julia
using Distributions, Random, LinearAlgebra, Plots
```

```julia
nd = Normal(0,1)
```

```julia
data = rand(nd,1000,2)
```

```julia
scatter(data[:,1],data[:,2], aspect_ratio = :equal, legend = nothing )
```

```julia
data[1,:]
```

```julia
datas = copy(data)
datas[:,2] = datas[:,2]*3
```

```julia
scatter(datas[:,1],datas[:,2], aspect_ratio = :equal, legend = nothing )
```

```julia
θ = π/4
```

```julia
r = [[cos(θ), -sin(θ)] [sin(θ), cos(θ)]] 
```

```julia
datar = datas * r
```

```julia
scatter(datar[:,1],datar[:,2], aspect_ratio = :equal, legend = nothing )
```

```julia
u,s,v = svd(datar/sqrt(1000))
```

```julia
v * r
```

```julia
delta = [1, 1]
datar_translated = datar .+ delta'

scatter(datar_translated[:, 1], datar_translated[:, 2], aspect_ratio=:equal, legend=nothing)
```

```julia
mean_x = mean(datar_translated[:, 1])
mean_y = mean(datar_translated[:, 2])

datar_centered = hcat(datar_translated[:, 1] .- mean_x, datar_translated[:, 2] .- mean_y)
```

```julia
U, S, V = svd(datar_centered)

datar_pca = datar_centered * V
```

```julia
data_centered = data .- mean(data, dims=1)

U, S, V = svd(data_centered)

function plot_ellipse(center, axes, rotation_matrix, sigma_level, color)
    θ = range(0, 2π, length=100) 
    x = cos.(θ) .* sqrt(axes[1]) * sigma_level 
    y = sin.(θ) .* sqrt(axes[2]) * sigma_level
    ellipse = hcat(x, y) * rotation_matrix  # Apply rotation to align with principal components
    plot!(ellipse[:, 1] .+ center[1], ellipse[:, 2] .+ center[2], lw=2, label="Ellipse at $(sigma_level)σ", color=color)
end

scatter(data_centered[:, 1], data_centered[:, 2], aspect_ratio=:equal, legend=false, title="Original Distribution with Ellipses")

center = [0, 0] 
sigma_levels = [1, 2, 3]  
colors = [:orange, :green, :purple] 
for (i, sigma_level) in enumerate(sigma_levels)
    plot_ellipse(center, S, V, sigma_level, colors[i]) 
end

quiver!([0], [0], quiver=([V[1, 1]] * sqrt(S[1]), [V[2, 1]] * sqrt(S[1])), color=:red, label="PC1", lw=4)
quiver!([0], [0], quiver=([V[1, 2]] * sqrt(S[2]), [V[2, 2]] * sqrt(S[2])), color=:green, label="PC2", lw=4)

xlims!(-18, 18)
ylims!(-18, 18)
```

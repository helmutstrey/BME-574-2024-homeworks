```julia
using Distributions, Random, LinearAlgebra, Plots
```

```julia
nd = Normal(0,1)
data = rand(nd,(1000,2))
```

```julia
scatter(data[:,1],data[:,2], aspect_ratio = :equal, legend = nothing )
```

```julia
data[1,:]
```

```julia
delta_x = 1
delta_y = 1
datat = data .+ [delta_x delta_y]
```

```julia
scatter(datat[:,1], datat[:,2], aspect_ratio=:equal, legend=nothing)
```

```julia
datas = copy(datat)
datas[:,2] = datas[:,2]*3
```

```julia
scatter(datas[:,1],datas[:,2], aspect_ratio = :equal, legend = nothing )
```

```julia
θ = π/4
r = [[cos(θ), -sin(θ)] [sin(θ), cos(θ)]] 
datar = datas * r
```

```julia
scatter(datar[:,1], datar[:,2], aspect_ratio=:equal, legend=nothing)
```

```julia
mean_x, mean_y = mean(datar, dims=1)
datac = datar .- [mean_x mean_y]
```

```julia
scatter(datac[:,1], datac[:,2], aspect_ratio=:equal, legend=nothing)
```

```julia
u,s,v = svd(datac/sqrt(1000))
```

```julia
cov_matrix = Diagonal(s.^2)
theta = range(0, stop = 2π, length = 100)
circle = [cos.(theta) sin.(theta)]
for sigma in 1:3
    ellipse_points = sigma * circle * sqrt(cov_matrix)
    scatter!(mean_x .+ ellipse_points[:, 1], mean_y .+ ellipse_points[:, 2], label = "$(sigma)-sigma ellipse", lw = 2)
end
```

```julia
principal_vectors = v * Diagonal(s)
plot!([0, principal_vectors[1, 1]], [0, principal_vectors[2, 1]], label = "1st Principal Component", lw = 2, color = :red)
plot!([0, principal_vectors[1, 2]], [0, principal_vectors[2, 2]], label = "2nd Principal Component", lw = 2, color = :blue)
scatter!(legend = :outertopright, aspect_ratio = :equal)
```

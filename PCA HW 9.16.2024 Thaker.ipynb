```julia
using Distributions, Random, LinearAlgebra, Plots
```

```julia
nd = Normal(0,1)
```

```julia
data = rand(nd, (1000,2))
```

```julia
datas = copy(data)
datas[:,2] = datas[:,2]*4
```

```julia
θ = π/4
```

```julia
r = [[cos(θ), -sin(θ)] [sin(θ), cos(θ)]] 
```

```julia
rot_data = datas*r
```

```julia
trans_data = rot_data .+ [1 1]
```

```julia
rowMeans = Vector{Float64}(undef, size(data, 1))

for i in 1:size(trans_data, 1)
    rowMeans[i] = mean(data[i,:])
end
```

```julia
B = trans_data .- rowMeans
```

```julia
u, s, vt = svd(B./sqrt(1000))
```

```julia
removeRot = trans_data*r'
```

```julia
pc = vt'
```

```julia
#trying to translate matlab example
rot_pc = pc * r
```

```julia
theta = range(0, stop = 2*pi, length = 1000)
circle = [cos.(theta) sin.(theta)] *r
std = circle * Diagonal(s) * vt'
xmean = mean(B, dims = 1)

sig1 = std .+ xmean
sig2 = 2*std .+ xmean
sig3 = 3*std .+ xmean

plot()
#scatter!(data[:, 1], data[:, 2], label = "Data Points")
#scatter!(removeRot[:, 1], removeRot[:, 2], label = "Data Points")
scatter!(trans_data[:, 1], trans_data[:, 2], label = "Data Points")
plot!(sig1[:, 1], sig1[:, 2], label = "1sigma")
plot!(sig2[:, 1], sig2[:, 2], label = "2sigma")
plot!(sig3[:, 1], sig3[:, 2], label = "3sigma")
quiver!([xmean[1]],[xmean[2]], quiver = ([vt'[1,1]], [vt'[2,1]]), label = "pc1", lw =3, color = :green)
quiver!([xmean[1]],[xmean[2]], quiver = ([vt'[1,2]], [vt'[2,2]]), label = "pc2", lw =3, color = :red)
scatter!([xmean[1]], [xmean[2]], label="PC1", color=:green, marker=:utriangle, markersize=3)
scatter!([xmean[1]], [xmean[2]], label="PC2", color=:red, marker=:utriangle, markersize=3)
```

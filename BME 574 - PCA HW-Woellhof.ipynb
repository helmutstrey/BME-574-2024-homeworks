```julia
using Distributions, Random, LinearAlgebra, Plots
```

```julia
nd = Normal(0,1)
```

```julia
# Generate Data

data = rand(nd,(1000,2))
```

```julia
scatter(data[:,1],data[:,2],aspect_ratio = :equal,legend = nothing)
```

```julia
matrix1 = copy(data)
```

```julia
# Stretch Data

matrix1[:,2] = matrix1[:,2]*3
```

```julia
scatter(matrix1[:,1],matrix1[:,2],aspect_ratio = :equal,legend = nothing)
```

```julia
# Rotate Data

theta = pi/4
```

```julia
r = [[cos(theta),-sin(theta)] [sin(theta),cos(theta)]]
```

```julia
datar = matrix1*r
```

```julia
scatter(datar[:,1],datar[:,2], aspect_ratio = :equal, legend = nothing )
```

```julia
# Translate Data
data_trans = datar .+ [1 1]
```

```julia
# Subtract Mean
dataMean = mean(data_trans,dims = 1)
data_cen = data_trans .- dataMean
```

```julia
# SVD

u,s,v =svd(data_cen/sqrt(1000))
```

```julia
scatter(data_trans[:,1],data_trans[:,2],label = "Translated Data", aspect_ratio = :equal)
```

```julia
# Circle Stuff

theta = range(0,2*pi,length = 1000)
data_st = [cos.(theta) sin.(theta)]*Diagonal(s)*v'
```

```julia
for i in range(1,3)
    plot!(dataMean[1] .+ data_st[:,1].*i,dataMean[2] .+ data_st[:,2].*i,
        label ="$(i) SD", lw = 2) # Plot details
end
```

```julia
# Principle Directions

pc1 = v[:,1]*(s[1])
pc2 = v[:,2]*(s[2])
```

```julia
plot!([dataMean[1],dataMean[1]+pc1[1]],[dataMean[2],dataMean[2]+pc1[2]], arrow = :arrow,linewidth = 2, label = "PC 1",color = :yellow)
plot!([dataMean[1],dataMean[1]+pc2[1]],[dataMean[2],dataMean[2]+pc2[2]],arrow = :arrow,linewidth = 2, label = "PC 2", color = :orange)
```

```julia
using Distributions, Random, LinearAlgebra, Plots
```

```julia
nd = Normal(0,1)
```

```julia
data = rand(nd,(1000,2))
```

```julia
scatter(data[:,1],data[:,2], aspect_ratio = :equal, legend = nothing )
```

```julia
data[1,:]
```

```julia
datas = copy(data)
datas[:,2] = datas[:,2]*3
```

```julia
scatter(datas[:,1],datas[:,2], aspect_ratio = :equal, legend = nothing )
```

```julia
θ = π/4
```

```julia
r = [[cos(θ), -sin(θ)] [sin(θ), cos(θ)]]
```

```julia
datar = datas * r
```

```julia
scatter(datar[:,1],datar[:,2], aspect_ratio = :equal, legend = nothing)
```

```julia
datat = datar .+ [1 1]
```

```julia
xmean = mean(datat[:,1])
```

```julia
ymean = mean(datat[:,2])
```

```julia
datasub = datat .- [xmean ymean]
```

```julia
u,s,v = svd(datasub/sqrt(1000))
```

```julia
scatter(datat[:,1],datat[:,2], aspect_ratio = :equal, label="Random Data")
```

```julia
θ1 = range(0, 2π, length = 1000) #range function found on jlHub Julia Forum
```

```julia
data_standard = [cos.(θ1) sin.(θ1)] * Diagonal(s) * v'
```

```julia
sigma1 = xmean .+ data_standard[:,1], ymean .+ data_standard[:,2]
sigma2 = xmean .+ data_standard[:,1].*2, ymean .+ data_standard[:,2].*2
sigma3 = xmean .+ data_standard[:,1].*3, ymean .+ data_standard[:,2].*3 
#Transcribed to Julia language from textbook example
```

```julia
plot!(sigma1, linewidth=2, color=:red, label=false)
plot!(sigma2, linewidth=2, color=:red, label=false)
plot!(sigma3, linewidth=2, color=:red, label="σ")
```

```julia
scale1 = v[:,1]*s[1]
#Scaling formula given in Figure 1.13b
```

```julia
scale2 = v[:,2]*s[2]
```

```julia
plot!([xmean, xmean+scale1[1, 1]], [ymean, ymean+scale1[2,1]], color=:yellow, linewidth=3, label=false)
plot!([xmean, xmean+scale2[1, 1]], [ymean, ymean+scale2[2,1]], color=:yellow, linewidth=3, label="Principal Direction")
```
